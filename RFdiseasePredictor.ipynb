```python
import os
import joblib
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler 

from sklearn.tree import export_graphviz
```

```python
# Loading in the Training Data
df = pd.read_csv('resources/Training.csv')
df.head()
```

```python
#le = preprocessing.LabelEncoder()
#df = df.apply(le.fit_transform)
```

```python
# Setting X and y variables X= symptoms , y = prognosis
X = df.iloc[:, 0:-1]
y= df['prognosis']
```

```python
# View shape of first few elements of the data

print(X[:5])
print(X.shape)
print(y[:5])
print(y.shape)
```

```python
# 4920 rows with 132 features per row.
```

```python
# Split data into Training and Validation data using 'train_test_split' function

X_train, X_test, y_train, y_test = train_test_split(X, y,)
```

```python
# Display the Shape

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Split = 75% Training Data & 25% Test Data
```

```python
# Setting split size parameter using 'test_size'
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
# Verifying New Split by Displaying Shape
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Split now = 70% Training & 30% Testing Data
```

```python
# By default a random set of data is used for the split
```

```python
# Feature Scaling (Standard Scaler)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# 'random_state' is used to seed the same subset of data of the split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=8)
```

```python
# Random Forest Classifier
randomForestClass = RandomForestClassifier()
```

```python
randomForestClass.fit(X_train, y_train)

print('Split Test Accuracy:' , randomForestClass.score(X_test, y_test))
```
