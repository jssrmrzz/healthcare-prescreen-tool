```python
import pandas as pd
import numpy as np
import tensorflow as tf

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
# Loading in the Training Data
df = pd.read_csv('resources/Training.csv')
df.head()
```

```python
X = df.iloc[:, 0:-1]
y= df['prognosis']
```

```python
# Setting split size parameter using 'test_size'
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
# Verifying New Split by Displaying Shape
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Split now = 70% Training & 30% Test Data
```

```python
# By default a random set of data is used for the split
```

```python
# 'random_state' is used to seed the same subset of data of the split
y = pd.get_dummies(df, columns=['prognosis'], drop_first=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=8)
```

```python
y.shape
```

```python
# Scaler
scaler = StandardScaler()
X_scaler = scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Nueral network
nn = tf.keras.models.Sequential()

# Layer 1
nn.add(tf.keras.layers.Dense(units=500, activation='relu', input_dim=len(X_train_scaled[0])))

# Layer 2
nn.add(tf.keras.layers.Dense(units=250, activation='relu'))

# Layer 3
nn.add(tf.keras.layers.Dense(units=270, activation='relu'))

# Layer 4
nn.add(tf.keras.layers.Dense(units=172, activation='sigmoid'))

nn.summary()
```

```python
# Compile the model
nn.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=200)
```

```python
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
